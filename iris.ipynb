```python
'''
petal_length > 4.5 & species = "Iris-virginica"
'''
df_iris.query("petal_length > 4.5 and species == 'Iris-virginica' ")
```

```python
# Average Sepal Length (grouped by species)

df_iris.groupby("species")["sepal_length"].mean()
```

```python
# Maximum Petal Width (grouped by species)

df_iris.groupby("species")["petal_width"].max()
```

```python
# Standard Deviation of Sepal Width (grouped by species) 

df_iris.groupby("species")["sepal_width"].std()
```

```python
# Combining all with aggregation 
df_iris.groupby("species").agg(
    mean_sepal_length = ("sepal_length", "mean"),
    max_petal_width = ("petal_width", "max"),
    std_sepal_width = ("sepal_width", "std")
)
```

```python
# New Column --> petal_ratio
df_iris = df_iris.assign(petal_ratio = df_iris["petal_length"]/df_iris["petal_width"])
```

```python
# Average petal ratio grouped by species
df_iris.groupby("species")["petal_ratio"].mean()
```
